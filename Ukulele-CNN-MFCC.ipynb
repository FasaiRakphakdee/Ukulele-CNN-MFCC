### Mel-frequency cepstral coefficients (MFCC) extraction
function returns:

```python
from python_speech_features import mfcc
import scipy.io.wavfile
import numpy as np


chords_num = ["A-01","A7-02","Am-03","Am7-04","B-05","B7-06","Bm-07","Bm7-08","C-09","C7-10","Cm-11","Cm7-12",
                "D-13","D7-14","Dm-15","Dm7-16","E-17","E7-18","Em-19","Em7-20","F-21","F7-22","Fm-23","Fm7-24",
                "G-25","G7-26","Gm-27","Gm7-28"]
BGs = ["01","02","03","05","04","06","07"]

def mfcc_extraction(path, modal_song, chords_num, BGs):
    labels = []
    mfcc_features = []
    
    #01-01-01
    for index1,chord_num in enumerate(chords_num):
        for index,bg in enumerate(BGs):
            sig, rate = librosa.load(path+modal_song+chord_num[-2:]+"-01-01-01-"+bg+".wav", sr=16000)
            mfcc_feat = mfcc(sig,rate)
            mfcc_features.append(mfcc_feat)
            
            labels.append(chord_num[:-3])
    #01-01-02
    for index1,chord_num in enumerate(chords_num):
        for index,bg in enumerate(BGs):
            sig, rate = librosa.load(path+modal_song+chord_num[-2:]+"-01-01-02-"+bg+".wav", sr=16000)
            mfcc_feat = mfcc(sig,rate)
            mfcc_features.append(mfcc_feat)
            
            labels.append(chord_num[:-3])
    #01-01-03
    for index1,chord_num in enumerate(chords_num):
        for index,bg in enumerate(BGs):
            sig, rate = librosa.load(path+modal_song+chord_num[-2:]+"-01-01-03-"+bg+".wav", sr=16000)
            mfcc_feat = mfcc(sig,rate)
            mfcc_features.append(mfcc_feat)
            
            labels.append(chord_num[:-3])
    #01-01-04
    for index1,chord_num in enumerate(chords_num):
        for index,bg in enumerate(BGs):
            sig, rate = librosa.load(path+modal_song+chord_num[-2:]+"-01-01-04-"+bg+".wav", sr=16000)
            mfcc_feat = mfcc(sig,rate)
            mfcc_features.append(mfcc_feat)
            
            labels.append(chord_num[:-3])
    #01-01-05
    for index1,chord_num in enumerate(chords_num):
        for index,bg in enumerate(BGs):
            sig, rate = librosa.load(path+modal_song+chord_num[-2:]+"-01-01-05-"+bg+".wav", sr=16000)
            mfcc_feat = mfcc(sig,rate)
            mfcc_features.append(mfcc_feat)
            
            labels.append(chord_num[:-3])
    #01-02-01
    for index1,chord_num in enumerate(chords_num):
        for index,bg in enumerate(BGs):
            sig, rate = librosa.load(path+modal_song+chord_num[-2:]+"-01-02-01-"+bg+".wav", sr=16000)
            mfcc_feat = mfcc(sig,rate)
            mfcc_features.append(mfcc_feat)
            
            labels.append(chord_num[:-3])
    #01-02-02
    for index1,chord_num in enumerate(chords_num):
        for index,bg in enumerate(BGs):
            sig, rate = librosa.load(path+modal_song+chord_num[-2:]+"-01-02-02-"+bg+".wav", sr=16000)
            mfcc_feat = mfcc(sig,rate)
            mfcc_features.append(mfcc_feat)
            
            labels.append(chord_num[:-3])
    #01-02-03
    for index1,chord_num in enumerate(chords_num):
        for index,bg in enumerate(BGs):
            sig, rate = librosa.load(path+modal_song+chord_num[-2:]+"-01-02-03-"+bg+".wav", sr=16000)
            mfcc_feat = mfcc(sig,rate)
            mfcc_features.append(mfcc_feat)
            
            labels.append(chord_num[:-3])
    #01-02-04
    for index1,chord_num in enumerate(chords_num):
        for index,bg in enumerate(BGs):
            sig, rate = librosa.load(path+modal_song+chord_num[-2:]+"-01-02-04-"+bg+".wav", sr=16000)
            mfcc_feat = mfcc(sig,rate)
            mfcc_features.append(mfcc_feat)
            
            labels.append(chord_num[:-3])      
    #01-02-05
    for index1,chord_num in enumerate(chords_num):
        for index,bg in enumerate(BGs):
            sig, rate = librosa.load(path+modal_song+chord_num[-2:]+"-01-02-05-"+bg+".wav", sr=16000)
            mfcc_feat = mfcc(sig,rate)
            mfcc_features.append(mfcc_feat)
            
            labels.append(chord_num[:-3])
            
    return mfcc_features, labels


mfcc_features_and_labels = mfcc_extraction(my_path,'/03-02-', chords_num, BGs)
```

# zero pad function

```python
def zero_pad(features):
    
    new_spoken_train = []
    max_shape = 0
    for x in features:
        all_shapes = x.shape[0]
        if all_shapes > max_shape:
            max_shape = all_shapes

    for x in features:
        difference = max_shape-x.shape[0]
        zero_padded = np.pad(x,((0,difference),(0,0)), "constant")
        new_spoken_train.append(zero_padded)

    new_spoken_train = np.array(new_spoken_train)
    
    return new_spoken_train

mfcc_features = zero_pad(mfcc_features)
```

## one-hot encode labels (X_train)

```python
from numpy import array
from numpy import argmax
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

# define example
y_train = y_train
y_train_values = array(y_train)
# integer encode
y_train_label_encoder = LabelEncoder()
y_train_integer_encoded = y_train_label_encoder.fit_transform(y_train_values)
# binary encode
y_train_onehot_encoder = OneHotEncoder(sparse=False)
y_train_integer_encoded = y_train_integer_encoded.reshape(len(y_train_integer_encoded), 1)
y_train_onehot_encoded = y_train_onehot_encoder.fit_transform(y_train_integer_encoded)
print(y_train_onehot_encoded.shape)
```

## one-hot encode labels (X_train)

```python
# define example
y_test = y_test
y_test_values = array(y_test)
# integer encode
y_test_label_encoder = LabelEncoder()
y_test_integer_encoded = y_test_label_encoder.fit_transform(y_test_values)
# binary encode
y_test_onehot_encoder = OneHotEncoder(sparse=False)
y_test_integer_encoded = y_test_integer_encoded.reshape(len(y_test_integer_encoded), 1)
y_test_onehot_encoded = y_test_onehot_encoder.fit_transform(y_test_integer_encoded)
print(y_test_onehot_encoded.shape)
```

# Convolutional Neural network

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Embedding
from keras.layers import Input, Flatten, Dropout, Activation
from keras.layers import Conv1D, MaxPooling1D
from keras.models import Model
from keras.callbacks import ModelCheckpoint

len_of_labels = len(y_train_onehot_encoded[0])
model = Sequential()

model.add(Conv1D(128, 4, padding='same', input_shape=(X_train.shape[1],X_train.shape[2])))
model.add(Activation('relu'))
model.add(MaxPooling1D(pool_size=(2)))
model.add(Conv1D(128, 4, padding='same',))
model.add(Activation('relu'))
model.add(Dropout(0.4))
model.add(MaxPooling1D(pool_size=(2)))
model.add(Conv1D(128, 4, padding='same',))
model.add(Activation('relu'))
model.add(Dropout(0.4))
model.add(MaxPooling1D(pool_size=(2)))
model.add(Dropout(0.4))
model.add(Flatten())
model.add(Dense(len_of_labels))
model.add(Activation('softmax'))
```

```python
model.summary()
```

```python
from keras.optimizers import RMSprop

model.compile(loss= "categorical_crossentropy",
              optimizer= RMSprop(lr=0.001),
              metrics=['accuracy'])

cnnhistory= model.fit(X_train, y_train_onehot_encoded,batch_size= 64, epochs= 40,validation_data=(X_test, y_test_onehot_encoded))
```

```python
result = model.evaluate(X_test, y_test_onehot_encoded)
print("validation accuracy:", result[1])
```
